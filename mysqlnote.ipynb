2. SHOW DATABASES
3. DROP DATABASE `mysql`

### DATA TYPE

1. INT
2. DECIMAL(m, n) : (m位數, n位小數) 
ex: DECIMAL(3, 2) : 3.25
3. VARCHAR(n)：最大n位字串
4. BLOB : binary large object
5. DATE 
6. TIMESTAMP

### TABLE

1. USE `mysqlhw`;
CREATE TABLE `student`(
`student_id` INT,
`name` VARCHAR(20),
`major` VARCHAR(20),
PRIMARY KEY(`student_id`)
);
2. DESCRIBE `student`
3. DROP TABLE `student`
4. ALTER TABLE `student` ADD `gpa` DECIMAL(3,2)
5. ALTER TABLE `student` DROP `gpa`

### UPDATE & DELETE

1. UPDATE `student`
SET `major` = ‘行銷’
WHERE `student_id` = ‘2’

WHERE `score` < 60

### SELECT

1. SELECT *
FROM `student`
